```python
%reload_ext autoreload
%autoreload 2
import os, sys
os.environ["CUDA_DEVICE_ORDER"]="PCI_BUS_ID"
os.environ["CUDA_VISIBLE_DEVICES"]="5"  # specify which GPU(s) to be used
import torch
import torch.nn as nn
import numpy as np
from tqdm import tqdm_notebook as tqdm
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import seaborn as sns
import pandas as pd
import time
import matplotlib.pyplot as plt
import pickle as pkl
```

```python
df = pd.read_csv("../Train_Data.csv")
df.head()
```

```python
fold = 0
percent = 0.1
df_train_fold = df.loc[df["split"] != fold]
np.random.seed(2020)
idxs = []
for k_means_grp in range(12):
    df_train_fold_k = df_train_fold.loc[df_train_fold["k_means_grp"] == k_means_grp]
    df_train_fold_k_c = df_train_fold_k.loc[df_train_fold_k["curicullum_cluster_grp"] == 0]
    idxs_to_choose = df_train_fold_k_c["image_idx"].to_numpy()
    idx_select = np.random.choice(idxs_to_choose, int(percent * len(df_train_fold_k)), replace = False)
    idxs.append(idx_select)
idxs = np.concatenate(idxs,0)
```

```python
np.save(f"curriculum_easy_10idx_{fold}.npy",idxs)
```

```python
df_select = df.loc[idxs]
```

```python
def plot_relative_distribution(df, feature, hue, title='', size=2):
    f, ax = plt.subplots(1,1, figsize=(4*size,3*size))
    total = float(len(df))
    sns.countplot(x=feature, hue=hue, data=df, palette='Set2')
    plt.title(title)
    for p in ax.patches:
        height = p.get_height()
        ax.text(p.get_x()+p.get_width()/2.,
                height,
                '{:1.2f}%'.format(100*height/total),
                ha="center") 
    plt.show()
```

```python
plot_relative_distribution(df=df_select, feature='isup_grade', hue='data_provider', 
                                   title = f'relative count plot of isup_grade with data provider.', 
                                   size=2)
plot_relative_distribution(df=df_select, feature='isup_grade', hue=None, 
                                   title = f'relative count plot of isup_grade with curriculum selection.', 
                                   size=2)
```

```python
plot_relative_distribution(df=df_train_fold, feature='isup_grade', hue='data_provider', 
                                   title = f'relative count plot of isup_grade with data provider.', 
                                   size=2)
plot_relative_distribution(df=df_train_fold, feature='isup_grade', hue=None, 
                                   title = f'relative count plot of isup_grade with data provider.', 
                                   size=2)
```
